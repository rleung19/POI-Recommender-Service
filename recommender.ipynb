```python
target_city_dict = {
    "1": {"id": "1", "name":"Edinburgh"},
    "2": {"id": "2", "name": "Las Vegas"}
}


for city_id in target_city_dict:
    target_city = target_city_dict[city_id]
    target_city_business_all = top_business[top_business.city == target_city["name"]]
    target_city_business  = target_city_business_all[\
                                                      target_city_business_all.categories.str.contains("Arts & Entertainment") |
                                                      target_city_business_all.categories.str.contains("Shopping") |
                                                      target_city_business_all.categories.str.contains("Fashion") |                                                              
                                                      target_city_business_all.categories.str.contains("Arts & Crafts") |                                                              
                                                      target_city_business_all.categories.str.contains("Landmarks & Historic Buildings") |                                                              
                                                      target_city_business_all.categories.str.contains("Active Life") |                                                              
                                                      target_city_business_all.categories.str.contains("Men's Clothing") |                                                              
                                                      target_city_business_all.categories.str.contains("Women's Clothing") |                                                              
                                                      target_city_business_all.categories.str.contains("Sports Wear") |                                                              
                                                      target_city_business_all.categories.str.contains("Sporting Good") |                                                              
                                                      target_city_business_all.categories.str.contains("Local Flavor") |                                                              
                                                      target_city_business_all.categories.str.contains("Music & Video") | 
                                                      target_city_business_all.categories.str.contains("Books") |                                                              
                                                      target_city_business_all.categories.str.contains("Performing Arts") |                                                              
                                                      target_city_business_all.categories.str.contains("Jewelry") |                                                              
                                                      target_city_business_all.categories.str.contains("Accessories") |                                                              
                                                      target_city_business_all.categories.str.contains("Home & Garden") |                                                                                                                          
                                                      target_city_business_all.categories.str.contains("Art Galleries")]

    target_city_business = target_city_business[target_city_business.categories.str.contains('Food|Grocery|Beauty Supply|Furniture|Home Decor|Martial Arts|Mattress|Tobacco|Vape|Local Service|Nightlife|Hair Salons|Health & Medical|Home Services|Event Planning|Auto|Financial|Professional|Hair|Lawyer|Skin Care|Education|Waxing|Eyelash|Waxing|Tattoo|Gyms|Oil Change|Personal Injury') == False]

    #produce an inner join for EDN only reviews
    target_city_review = pd.merge(review, target_city_business, on='business_id', how='inner')
    
    # merge categories and review text
    target_city_review['text_categories'] =  target_city_review['categories'] + '; ' \
    + target_city_review['text']
    

    # create a feature vectors by vectorizing the corpus into a CountVecors of 
    # unigrams and bigrams.
    # trigrams are not used because it will consume a lot of resources
    vectorizer=CountVectorizer(ngram_range=(1,2), max_df=0.5, min_df=0.01) 
    X_train_counts=vectorizer.fit_transform(target_city_review.text_categories)

    # the CountVectors are converted into TF-IDF scores
    tfidf_transformer = TfidfTransformer()
    X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)

    # Train a Multinomial Classifier using the TF-IDF Feature Vectors
    clf = MultinomialNB().fit(X_train_tfidf, target_city_review.business_id)

    target_city["reviews"] = target_city_review
    target_city["trained_model"] = clf
    target_city["vectorizer"] = vectorizer  
    target_city["target_city_business"] = target_city_business  
    
    
    
```
